## **Indexación y segmentación**

Se puede cambiar el elemento de una matriz utilizando la notación con corchetes y el índice asociado:


```python
c = np.array([20,1,2,3,4])
c
```

```python
c[0] = 100
c
```

```python
c[4] = 0
c
```

Al igualque las listas y las tuplas, se puede cortar una matriz numpy. Con el ejemplo de *c* se pueden asignar los elementos de 1 al 3 a una nueva matriz:

```python
d = c[1:4]
d
```

*Nota:* Al igual que las listas, no se copia el elemento correspondiente al último índice

Se puede asignar los índices correspondientes a nuevos valores de la siguiente manera:

```python
c[3:5] = 300,400
c
```

## **Operaciones básicas**

 *Numpy hace que sea más fácil hacer muchas operaciones que se realizan comúnmente en la ciencia de datos. Estas mismas operaciones suelen ser computacionalmente más rápidas y requieren menos memoria en comparación con
a Python normal.*

### Adición y sustracción de vectores

Laadición y sustracción se realizan componente a componente. El resultado es una combinación lineal de ambos vectores, por ejemplo:

```python
u = np.array([1,0])
v = np.array([0,1])
z = u + v
z
```

```python
u = np.array([5,4])
v = np.array([7,1])
z = u - v
z
```

### Multiplicación de arreglos con un escalar

Cada componente del vector se multiplica por el escalar, por ejemplo:

```python
y = np.array([1,2])
z = 2*y
z
```

### Multiplicación de dos arreglos numpy

La multiplicación de vectores se hace componente a componente, el primer componente del vector resultante es el producto del primer elemento de cada vector que se multiplica. Ejemplo:

```python
u = np.array([1,2])
v = np.array([3,2])
z = u*v
z
```

### Producto punto

El producto punto representa cuán similares son los dos vectores. Para realizar el producto punto entre dos vectores se utiliza la función *dot* de numpy, por ejemplo:

```python
u = np.array([1,2])
v = np.array([3,1])
result = np.dot(u,v)
result
```

### Suma de constantes con un arreglo numpy

Si se agrega un valor escalar a la matriz, numpy agregará ese valor a cada elemento. Esta propiedad se conoce como broadcasting. Ejemplo:

```python
u = np.array([1,2,3,-1])
z = u + 1
z
```

## **Funciones universales**

Una función universal es una función que opera en ndarrays

* mean(): media o valor medio de todos los elementos

```python
a = np.array([1,-1,1,-1])
mean_a = a.mean()
mean_a
```

* max(): Devuelve el valor máximo del arreglo

```python
b = np.array([1,-2,3,4,5])
max_b = b.max()
max_b
```

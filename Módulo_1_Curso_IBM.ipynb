#**Tema 2: Comentarios en Python**

### **Uso de comentarios**

Los comentarios se pueden usar para:

1. Explicar el código Python.

2. Hacer que el código sea más legible.

3. Evitar la ejecución al probar el código.

### Creación de un comentario

Los comentarios comienzan con un # y Python ignorará als lineas siguientes.

```python
#This is a comment
print("Hello, World!")
```

Los comentarios se pueden colocar al final de una línea:

```python
print("Hello, World!") #This is a comment
```

Un comentario no tiene que ser texto que explique el código, también se puede usar para evitar que Python ejecute código:

```python
#print("Hello, World!")
print("Rico, ¡Aguardiente!")
```

### Comentarios multilínea


Python realmente no tiene una sintaxis para comentarios de varias líneas. Para agregar un comentario multilínea, puede insertar un # para cada línea

```python
#This is a comment
#written in
#more than just one line
print("Hello, World!")
```

O, dado que Python ignorará los literales de cadena que no están asignados a una variable, se puede agregar una cadena multilínea (comillas triples) en el código y colocar el comentario dentro de él:

```python
"""
This is a comment
written in
more than just one line
"""
print("Hello, World!")

# Mientras la cadena no esté asignada a una variable, Python leerá el código, pero luego lo ignorará y habrá hecho un comentario multilínea.
```

En Python, el tipo de datos se establece cuando se asigna un valor a una variable:

```python
from IPython.core.display import Javascript
A = "Hello World"	#str	
B = 20	#int	
C = 20.5	#float	
D = 1j	#complex	
E = ["apple", "banana", "cherry"]	#list	
F = ("apple", "banana", "cherry")	#tuple	
G = range(6)	#range	
H = {"name" : "John", "age" : 36}	#dict	
I = {"apple", "banana", "cherry"}	#set	
J = frozenset({"apple", "banana", "cherry"})	#frozenset	
K = True	#bool	
L = b"Hello"	#bytes	
# M = bytearray(5)	#bytearray	
# N = memoryview(bytes(5))	#memoryview	
O = None	#NoneType

print(A)
print(B)
print(C)
print(D)
print(E)
print(F)
print(G)
print(H)
print(I)
print(J)
print(K)
print(L)
#print(M)
#print(N)
print(O)
```

### Especificar un tipo de variable

Puede haber ocasiones en las que desee especificar un tipo en una variable. Esto se puede hacer con casting. Python es un lenguaje orientado a objetos y, como tal, utiliza clases para definir tipos de datos, incluidos sus tipos primitivos.

Por lo tanto, la fundición en Python se realiza utilizando funciones de constructor:

* int(): construye un número entero a partir de un literal entero, un literal flotante (eliminando todos los decimales) o un literal de cadena (siempre que la cadena represente un número entero)

* float() - construye un número float a partir de un literal entero, un literal float o un literal string (siempre que la cadena represente un float o un entero)

* str() - construye una cadena a partir de una amplia variedad de tipos de datos, incluyendo cadenas, literales enteros y literales flotantes

```python
x = int(1)   # x will be 1
y = int(2.8) # y will be 2
z = int("3") # z will be 3

print (x, y, z)
```

```python
x = float(1)     # x will be 1.0
y = float(2.8)   # y will be 2.8
z = float("3")   # z will be 3.0
w = float("4.2") # w will be 4.2

print (x, y, z, w)
```

```python
x = str("s1") # x will be 's1'
y = str(2)    # y will be '2'
z = str(3.0)  # z will be '3.0'

print (x, y, z)
```

#**Tema 4: Cadenas de caracteres**

### **Generalidades**

Las cadenas en python están rodeadas por comillas simples o comillas dobles: 'hola' es lo mismo que "hola".

```python
print("Hello")
print('Hello')
```

La asignación de una cadena a una variable se realiza con el nombre de la variable seguido de un signo igual y la cadena:

```python
a = "Mi nombre es Henry"
print(a)
```

Se puede asignar una cadena multilínea a una variable mediante tres comillas, simples o dobles:

```python
a = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""
print(a)
```

### **Las cadenas son matrices**

Al igual que muchos otros lenguajes de programación populares, las cadenas en Python son matrices de bytes que representan caracteres unicos.

```python
a = "Hello, World!"
print(a[1])
print(a[3])
```

Dado que las cadenas son matrices, podemos recorrer los caracteres de una cadena, con un bucle for. 

```python
for x in "Colombia":
  print(x)
```

La asignación de una cadena a una variable se realiza con el nombre de la variable seguido de un signo igual y la cadena:

```python
a = "Mi nombre es Henry"
print(a)
```

Se puede asignar una cadena multilínea a una variable mediante tres comillas, simples o dobles:

```python
a = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""
print(a)
```

### **Longitud de la cadena**

Para obtener la longitud de una cadena, se utiliza la función len()

```python
a = "Hello, World!"
print(len(a))
```

### **Extracción de conjunto de caracteres**

Se puede devolver un rango de caracteres mediante la sintaxis de división (:)Así:



1.   Definiendo posición inicial y final Variable[2:5]
2.   Definiendo solo posición final  Variable[:5]
3. Hasta el final  Variable[2:]
4. De atras hacia adelante (Indexación Negativa)  Variable[-5:-2]



```python
b = "Hello, World!"
print(b[2:5]) #NO tiene en cuenta el ultimo digito
```

```python
b = "Hello, World!"
print(b[:5])
```

```python
b = "Hello, World!"
print(b[2:])
```

```python
b = "Hello, World!"
print(b[-5:-2])
```

##**Modificaciones de la cadena**

Python tiene un conjunto de métodos integrados que se puede usar en cadenas.

### Mayúscula

El método que devuelve la cadena en mayúsculas es :upper()

```python
a = "Hello, World!"
print(a.upper())
```

### Minúsculas

El método que devuelve la cadena en Minúsculas es :lower()

```python
a = "HELLO, WORLD!"
print(a.lower())
```

### Eliminar espacios en blanco

El método elimina cualquier espacio en blanco del principio o del final:strip()

```python
a = " Hello, World! "
print(a.strip()) # returns "Hello, World!"
```

### Reemplazar cadena

El método reemplaza una cadena por otra cadena:replace()

```python
a = "Hello, World!"
print(a.replace("H", "J"))
```

### Cadena dividida

El método divide la cadena en subcadenas si encuentra instancias del separador:split()

```python
a = "Hello, World!"
print(a.split(",")) # Retorna ['Hello', ' World!']
```

### **Comprobación de existencia de un caracter**

Para comprobar si una determinada frase o carácter está presente en una cadena, podemos utilizar la palabra clave .in

```python
a = "The best things in life are free!"
print("free" in a)

#Como resultado se tiene un dato de tipo bool
```

Para comprobar si una determinada frase o carácter NO está presente en una cadena, podemos utilizar la palabra clave .not in

```python
a = "The best things in life are free!"
print("expensive" not in a)
```

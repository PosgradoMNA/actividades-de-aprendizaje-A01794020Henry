#**Laboratorio**

```python
#Importanción de bibliotecas y librerias (Asignación de abreviaturas)

import pandas as pd 
import numpy as np
```

```python
# Traemos la base de datos de tipo CSV
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv" #Marcamos la direccion de nuestro dataset
df = pd.read_csv(path, header=None) #leemos el dataset

df
```

```python
# Imprimimos unicamente las primeras 5 filas de la base de datos
print("Las primeras cinco filas de la base de datos son") 
df.head(5) 
```

```python
# Se imprimen unicamente las ultimas 10 columnas (el valor numerico se sustituye por el numero de filas a imprimir)

print("Las ultimas 10 filas de la base de datos son:\n") 
df.tail(10)
```

```python
# A continuación se crean encabezados pertinetes para la información de la base de datos

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
# Luego de tener los encabezados, podemos reemplazarlos en la base de datos con el siguiente comando: 
df.columns = headers
df.head(20) #Se imprimen solo las primeras 20 filas
```

```python
#Podemos reemplazar valores usando la función df.replace
df1=df.replace('?',np.NaN) #Reemplazamos el simolo "?" por la expresion NaN

df1.head(20) #Se imprimen solo las primeras 20 filas
```

```python
#Con la función .dropna se puede eliminar los valores faltantes 

df=df1.dropna(subset=["price"], axis=0)
df.head(20)
```

```python
#Si la base de datos ya incluye los encabezados podemos saber el nombre de cada columna con la función .columns

print(df.columns) 
```

```python
# De igual manera se puede saber el tipo de dato que maneja cada columna con la función .dtypes

print(df.dtypes)
```

```python
#Con este comando obtenemos valores de relevancia para el analisis de los datos

df.describe() 
```

```python
# Para obtener una descripción general de la base de datos se usa la función .describe

df.describe(include = "all") 
```

```python
# También se pueden analizar unicamente las columnas seleccionadas

df[['length', 'compression-ratio']].describe() # Depende de la calidad de los datos
```

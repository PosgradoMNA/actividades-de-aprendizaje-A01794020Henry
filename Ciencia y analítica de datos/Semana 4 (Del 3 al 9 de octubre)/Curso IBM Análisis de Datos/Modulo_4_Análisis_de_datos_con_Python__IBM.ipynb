#**Tema 2: Regresión lineal**

El término regresión se utiliza cuando se intenta encontrar la relación entre las variables.

En el aprendizaje automático y en el modelado estadístico, esa relación se usa para predecir el resultado de eventos futuros.

image.png

La regresión lineal se refiere a como se relaciona una variable independiente con una dependiente y nos sirve para hacer una predicción. Para poder hacer predicciones usando la regresión lineal simple, debemos de seguir los siguientes pasos:

* Tener una variable predictora (variable idependiente) -x
* Tener un objetivo (variable dependiente) variable - y
* Encontrar la relacion lineal entre ambas variables

### **¿Como fucniona?**

La regresión lineal utiliza la relación entre los puntos de datos para dibujar una línea recta a través de todos ellos.

Esta línea se puede utilizar para predecir valores futuros.

image.png


En Machine Learning, predecir el futuro es muy importante.

Python tiene métodos para encontrar una relación entre puntos de datos y para dibujar una línea de regresión lineal. Te lo mostraremos Cómo usar estos métodos en lugar de pasar por la fórmula matemática.



```python
# Ejemplo
# CDibujando un diagrama de dispersión:

import matplotlib.pyplot as plt

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)
plt.show()
```

Importar y dibujar la línea de Regresión Lineal:scipy

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

#**Tema 4: Regresión polinómica**

Si sus puntos de datos claramente no se ajustarán a una regresión lineal (una línea recta a través de todos los puntos de datos), podría ser ideal para la regresión polinómica.

La regresión polinómica, como la regresión lineal, utiliza la relación entre el variables x e y para encontrar la mejor manera de trazar una línea a través de los puntos de datos.

image.png

### **¿Cómo funciona?**

Python tiene métodos para encontrar una relación entre puntos de datos y para dibujar Una línea de regresión polinómica. Le mostraremos cómo usar estos métodos en lugar de pasar por la fórmula matemática.

En el siguiente ejemplo, hemos registrado 18 coches a medida que pasaban un Ciertas casetas de peaje.

Hemos registrado la velocidad del coche, y la hora del día (hora) el paso Ocurrió.

El eje x representa las horas del día y el eje y representa el velocidad:

```python
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

plt.scatter(x, y)
plt.show()
```

Importar ya continuación, dibujar la línea de Regresión polinómica:numpy matplotlib

```python
import numpy
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

myline = numpy.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

### **R-cuadrado**

Es importante saber qué tan bien la relación entre los valores de la Los ejes x e y es, si no hay relación el polinomio La regresión no se puede utilizar para predecir nada.

La relación se mide con un valor llamado r cuadrado.

El valor de r cuadrado varía de 0 a 1, donde 0 significa que no hay relación, y 1 significa 100% relacionado.

Python y el módulo Sklearn calcularán este valor.

```python
# ¿Qué tan bien encajan mis datos en una regresión polinómica?

import numpy
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

print(r2_score(y, mymodel(x)))
```
